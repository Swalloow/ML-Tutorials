```python
import pandas as pd
import json
```

```python
FILE1 = '../Dataset/shopping1.json'
FILE2 = '../Dataset/shopping2.json'
FILE3 = '../Dataset/shopping3.json'
```

```python
file_list = [FILE1, FILE2, FILE3]
```

```python
with open(FILE1) as data_file:
    data1 = json.load(data_file)
with open(FILE2) as data_file:
    data2 = json.load(data_file)
with open(FILE3) as data_file:
    data3 = json.load(data_file)
```

```python
df1 = pd.DataFrame(data1['cont'])
df2 = pd.DataFrame(data2['cont'])
df3 = pd.DataFrame(data3['cont'])
df1['category'] = '쇼핑/서유럽'
df2['category'] = '쇼핑/스페인'
df3['category'] = '쇼핑/두바이'
```

```python
frames = [df1, df2, df3]
result = pd.concat(frames)
```
